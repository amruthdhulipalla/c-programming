```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([85,89,86,87,84,85,89])
y = np.array([4,5,6,2,3,1,7])

plt.plot(x, y)
plt.title("academics")
plt.xlabel("marks of the subjects")
plt.ylabel("no of attempts")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([85,89,86,87,84,85,89])
y = np.array([4,5,6,2,3,1,7])
font1={'color':'hotpink','size':'90'}
plt.plot(x, y)
plt.title("academics",fontdict=font1)
plt.xlabel("marks of the subjects",fontdict=font1)
plt.ylabel("no of attempts")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([85,89,86,87,84,85,89])
y = np.array([4,5,6,2,3,1,7])
font11={'size':'50','color':'g'}
plt.plot(x, y)
plt.title("academics",loc='right',fontdict=font11)
plt.xlabel("marks of the subjects",)
plt.ylabel("no of attempts",fontdict=font11)

plt.show()
```
